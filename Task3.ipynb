**Data visualization**

***Heatmap***

```
sns.heatmap(data.corr())#heatmap of correlation matrix
```

***Dark shades represents positive correlation while lighter shades represents negative correlation.***

```
sns.heatmap(data.corr(), annot = True)
```

***setting annot=True, well get values by which features are correlated to each other in grid-cells.***

***Boxplot***

```
sns.boxplot(x='Category', y='Profit', data=data)
```

```
sns.displot(data, x="Category",hue="Quantity")#hue will draw a separate histogram for each of its unique values.
```

***From the plot office supplies category is more in number***

***Univariate distributions: distributions of a single variable***

***Kernel Density Estimation***

```
sns.displot(data, x="Profit", kind="kde")#plotted to know the shape of the distribution
```

```
sns.displot(data, x="Profit",hue="Category", kind="kde")#kde for unique category
```

```
sns.displot(data, x="Quantity",hue="Category", kind="kde",multiple="stack")
#separate density estimate will be computed for each level of that hue variable
```

```
sns.displot(data, x="Quantity",hue="Sub-Category", kind="kde",fill="tree")
#distribution for quantity of each item in sub-category
```

***Bivariate distributions***

```
sns.displot(data, x="Quantity",y="Discount",hue="Category", kind="kde")#distributions of a two variable
```
